```python
import os
import glob
import matplotlib.pyplot as plt
import pandas as pd

# 1. Zoek naar ALLE labels in de hele content map (train + valid + test)
# We zoeken specifiek naar .txt files in mappen genaamd 'labels'
search_path = "/content/**/labels/*.txt"
label_files = glob.glob(search_path, recursive=True)

if not label_files:
    print("Geen labels gevonden. Controleer of de map /content/content/ bestaat.")
else:
    print(f"Totaal {len(label_files)} label-bestanden gevonden (train + valid)!")

    # Klasses uit je data.yaml
    class_names = ['nutri_A', 'nutri_B', 'nutri_C', 'nutri_D', 'shelf_bottom', 'shelf_middle', 'shelf_top']
    data = []

    for label_file in label_files:
        with open(label_file, 'r') as f:
            lines = f.readlines()

        shelves = []
        items = []
        for line in lines:
            parts = line.split()
            if not parts: continue
            cls, x, y, w, h = map(float, parts)
            name = class_names[int(cls)]

            if 'shelf' in name:
                # We slaan de boven- en onderkant van de plank op (Y-as)
                shelves.append({'name': name, 'y_min': y - h/2, 'y_max': y + h/2})
            elif 'nutri' in name:
                # We slaan het midden van het nutri-label op
                items.append({'name': name, 'y_center': y})

        # Koppel elk gevonden nutri-label aan de juiste plank
        for item in items:
            assigned_shelf = "Niet op plank"
            for shelf in shelves:
                if shelf['y_min'] <= item['y_center'] <= shelf['y_max']:
                    assigned_shelf = shelf['name']
                    break
            data.append({'Nutri-score': item['name'], 'Plank': assigned_shelf})

    # 2. Tabel maken en visualiseren
    if data:
        df_all = pd.DataFrame(data)

        # Zet de planken in een logische volgorde voor de grafiek
        plank_order = ['shelf_top', 'shelf_middle', 'shelf_bottom']
        df_all['Plank'] = pd.Categorical(df_all['Plank'], categories=plank_order, ordered=True)

        # Maak een kruistabel (pivot)
        pivot_df = df_all.groupby(['Plank', 'Nutri-score'], observed=False).size().unstack(fill_value=0)

        # 3. Grafiek tekenen
        # Kleuren passend bij Nutri-scores (Groen naar Rood)
        color_map = {'nutri_A': '#2ca02c', 'nutri_B': '#bcbd22', 'nutri_C': '#ff7f0e', 'nutri_D': '#d62728'}
        available_colors = [color_map[col] for col in pivot_df.columns if col in color_map]

        ax = pivot_df.plot(kind='bar', stacked=True, figsize=(12, 7), color=available_colors)

        plt.title('Totale Schap-analyse: Nutri-scores per Hoogte (Volledige Dataset)', fontsize=14)
        plt.ylabel('Totaal aantal labels gevonden')
        plt.xlabel('Locatie in het schap')
        plt.xticks(rotation=0)
        plt.legend(title="Nutri-score", bbox_to_anchor=(1.05, 1), loc='upper left')
        plt.grid(axis='y', linestyle='--', alpha=0.3)

        # Voeg aantallen toe bovenop de balken
        for rect in ax.patches:
            height = rect.get_height()
            if height > 0:
                ax.annotate(f'{int(height)}',
                            xy=(rect.get_x() + rect.get_width() / 2, rect.get_y() + height/2),
                            ha='center', va='center', fontsize=9, color='white', fontweight='bold')

        plt.tight_layout()
        plt.savefig('volledige_dataset_analyse.png')
        plt.show()

        print("\n--- Statistieken voor je bachelorproef ---")
        print(f"Totaal aantal geanalyseerde producten: {len(df_all)}")
        print(pivot_df)
    else:
        print("Geen combinaties van planken en nutri-scores gevonden. Staan de labels wel in YOLO formaat?")
```

```python
import os
import glob
import matplotlib.pyplot as plt
import pandas as pd

# 1. Data verzamelen uit alle mappen
# We zoeken in /content/ en alle submappen naar de labels
search_path = "/content/**/labels/*.txt"
label_files = glob.glob(search_path, recursive=True)

class_names = ['nutri_A', 'nutri_B', 'nutri_C', 'nutri_D', 'shelf_bottom', 'shelf_middle', 'shelf_top']
data = []

for label_file in label_files:
    with open(label_file, 'r') as f:
        lines = f.readlines()
    shelves = []
    items = []
    for line in lines:
        parts = line.split()
        if not parts: continue
        cls, x, y, w, h = map(float, parts)
        name = class_names[int(cls)]
        if 'shelf' in name:
            shelves.append({'name': name, 'y_min': y - h/2, 'y_max': y + h/2})
        elif 'nutri' in name:
            items.append({'name': name, 'y_center': y})

    for item in items:
        assigned_shelf = "Niet op plank"
        for shelf in shelves:
            if shelf['y_min'] <= item['y_center'] <= shelf['y_max']:
                assigned_shelf = shelf['name']
                break
        data.append({'Nutri-score': item['name'], 'Plank': assigned_shelf})

if not data:
    print("Geen data gevonden. Controleer of de labels in /content staan.")
else:
    df = pd.DataFrame(data)

    # --- De juiste volgorde instellen ---
    nutri_volgorde = ['nutri_A', 'nutri_B', 'nutri_C', 'nutri_D']
    plank_volgorde = ['shelf_top', 'shelf_middle', 'shelf_bottom']

    # Maak categoriën aan om de volgorde (A naar D en Top naar Bottom) te bewaren
    df['Nutri-score'] = pd.Categorical(df['Nutri-score'], categories=nutri_volgorde, ordered=True)

    # Filter op producten die op een gedefinieerde plank staan
    df_shelf = df[df['Plank'].isin(plank_volgorde)].copy()
    df_shelf['Plank'] = pd.Categorical(df_shelf['Plank'], categories=plank_volgorde, ordered=True)

    # 2. Berekeningen (Crosstab zonder 'observed')
    # Samenstelling per plank: "Welke Nutri-scores staan op deze plank?"
    shelf_comp = pd.crosstab(df_shelf['Plank'], df_shelf['Nutri-score'], normalize='index') * 100
    # Waar staat elke score: "Op welke planken vind ik deze specifieke Nutri-score?"
    score_dist = pd.crosstab(df_shelf['Nutri-score'], df_shelf['Plank'], normalize='index') * 100

    # 3. Visualisatie
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 14))
    nutri_colors = ['#2ca02c', '#bcbd22', '#ff7f0e', '#d62728'] # Groen(A) -> Rood(D)

    # --- Grafiek 1: Samenstelling per schaphoogte ---
    shelf_comp.plot(kind='barh', stacked=True, ax=ax1, color=nutri_colors)
    ax1.set_title('Samenstelling per schaphoogte (A → D)', fontsize=16, fontweight='bold', pad=20)
    ax1.set_xlabel('Percentage (%) van alle producten op deze plank', fontsize=12)
    ax1.invert_yaxis() # Top plank bovenaan
    ax1.legend(title="Nutri-score", labels=['A', 'B', 'C', 'D'], bbox_to_anchor=(1.0, 1))

    # --- Grafiek 2: Locatie per Nutri-score ---
    plank_colors = ['#3498db', '#9b59b6', '#f1c40f'] # Blauw(Top), Paars(Mid), Geel(Bot)
    score_dist.plot(kind='barh', stacked=True, ax=ax2, color=plank_colors)
    ax2.set_title('Waar staan de verschillende Nutri-scores? (A → D)', fontsize=16, fontweight='bold', pad=20)
    ax2.set_xlabel('Percentage (%) van alle producten met deze score', fontsize=12)
    ax2.set_yticklabels(['Nutri-score A', 'Nutri-score B', 'Nutri-score C', 'Nutri-score D'])
    ax2.invert_yaxis() # Nutri-A bovenaan
    ax2.legend(title="Plank Locatie", labels=['Bovenste plank', 'Middelste plank', 'Onderste plank'], bbox_to_anchor=(1.0, 1))

    # Voeg percentages toe in de balken
    for ax in [ax1, ax2]:
        for p in ax.patches:
            width = p.get_width()
            if width > 3: # Toon alleen tekst als de balk breed genoeg is
                ax.annotate(f'{width:.1f}%',
                            (p.get_x() + width/2, p.get_y() + p.get_height()/2),
                            ha='center', va='center', color='white', fontsize=10, fontweight='bold')

    plt.tight_layout()
    plt.savefig('resultaten_percentages_definitief.png', dpi=300)
    plt.show()

    print("\n✅ Analyse voltooid! Gebruik de tabel hieronder voor exacte cijfers:")
    print("\nSamenstelling per plank (%):")
    print(shelf_comp.round(1))
    print("\nLocatie per Nutri-score (%):")
    print(score_dist.round(1))
```

Diagrams of trained model

```python
import os
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# 1. Pad naar de map waar je resultaten staan
# Pas dit aan als je een andere 'name' hebt gebruikt bij het trainen
results_path = '/content/content/runs/detect/bachelor_yolo'

# 2. Welke bestanden willen we in de collage?
# We kiezen voor de voorspellingen, de confusion matrix en de PR-curve
files_to_find = [
    'val_batch0_pred.jpg',          # Foto's met voorspelde boxes
    'confusion_matrix_normalized.png', # Hoe goed de klasses herkend worden
    'BoxPR_curve.png',              # De Precision-Recall curve
    'results.png'                   # De algemene trainingsgrafieken
]

# Zoek de volledige paden op
existing_images = []
for f in files_to_find:
    full_path = os.path.join(results_path, f)
    if os.path.exists(full_path):
        existing_images.append(full_path)

# Voeg eventueel de extra grafieken toe die we eerder hebben gemaakt
extra_graphs = ['training_metrics.png', 'class_distribution.png', 'resultaten_A_naar_D.png']
for g in extra_graphs:
    if os.path.exists(g):
        existing_images.append(g)

if not existing_images:
    print("❌ Geen resultaat-afbeeldingen gevonden in de opgegeven map.")
else:
    # 3. Collage maken
    num_imgs = len(existing_images)
    cols = 2
    rows = (num_imgs + 1) // cols

    fig, axes = plt.subplots(rows, cols, figsize=(16, 6 * rows))
    fig.suptitle('Overzicht Modelresultaten Bachelorproef', fontsize=20, fontweight='bold', y=1.02)
    axes = axes.flatten()

    for i, img_path in enumerate(existing_images):
        img = mpimg.imread(img_path)
        axes[i].imshow(img)
        # Maak een mooie titel van de bestandsnaam
        title = os.path.basename(img_path).split('.')[0].replace('_', ' ').title()
        axes[i].set_title(title, fontsize=14, pad=10)
        axes[i].axis('off')

    # Verberg lege vakjes
    for j in range(i + 1, len(axes)):
        axes[j].axis('off')

    plt.tight_layout()
    plt.savefig('bachelorproef_collage_totaal.png', bbox_inches='tight', dpi=300)
    plt.show()

    print(f"✅ Collage is opgeslagen als 'bachelorproef_collage_totaal.png'")
```
